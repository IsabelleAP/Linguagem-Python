```python
#ordenando arquivos multi-fasta
from Bio import SeqIO
from functools import*

def comp(x,y):
  return len(x) - len(y)
seqs = list(SeqIO.parse('multifasta.fasta','fasta'))

#seqs.sort(cmp=lambda x,y: cmp(len(x), len(y)))
seqs.sort(key=cmp_to_key(comp))
SeqIO.write(seqs,'multifasta_sorted.fasta','fasta')
```

```python
import random

pop_size, seq_size, max_gen = 50, 20, 1000
cross_rate, mut_cross = 0.8, 0.1
pop = []

# Generate a random DNA sequence
def random_sequence(seq_size):
    return ''.join(random.choice('ACGT') for _ in range(seq_size))

# Calculate GC content
def gc_content(seq):
    return (seq.count('G') + seq.count('C')) / len(seq)

# Create initial population
for _ in range(pop_size):
    seq = random_sequence(seq_size)
    pop.append(seq)

# Genetic algorithm operations (crossover, mutation, selection) can be added here.

# Example: To maximize GC content, you can define selection, crossover, and mutation strategies.

# A function to select individuals based on fitness (GC content)
def select_population(population, num_selected):
    return sorted(population, key=gc_content, reverse=True)[:num_selected]

# A function for single-point crossover
def single_point_crossover(parent1, parent2):
    crossover_point = random.randint(1, len(parent1) - 1)
    child1 = parent1[:crossover_point] + parent2[crossover_point:]
    child2 = parent2[:crossover_point] + parent1[crossover_point:]
    return child1, child2

# A function for single-position mutation
def single_position_mutation(individual, mutation_rate):
    mutated_seq = list(individual)
    if random.random() < mutation_rate:
        mutation_point = random.randint(0, len(individual) - 1)
        mutated_seq[mutation_point] = random.choice('ACGT')
    return ''.join(mutated_seq)

# Evolution loop
for gen in range(max_gen):
    # Select the top-performing individuals
    selected_pop = select_population(pop, pop_size)

    # Perform crossover and mutation on selected individuals
    new_pop = []
    while len(new_pop) < pop_size:
        parent1, parent2 = random.choices(selected_pop, k=2)
        child1, child2 = single_point_crossover(parent1, parent2)
        child1 = single_position_mutation(child1, mut_cross)
        child2 = single_position_mutation(child2, mut_cross)
        new_pop.extend([child1, child2])

    # Replace the old population with the new one
    pop = new_pop

# Show the final population
print(pop)
```
