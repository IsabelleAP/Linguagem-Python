1. abrir o documento de microdados


```python
import pandas as pd
import numpy as np
```

```python
microdadossp = pd.read_csv('/content/microdados_enem_2019_sp.csv', sep=';', encoding='iso-8859-1')
```

```python
microdadossp.head()
```

2. verificar todos os tipos de atributos/ ver todas as colunas



```python
microdadossp.dtypes
```

3. deletar as colunas que não serão úteis

```python
dados = microdadossp.drop(columns=['CO_MUNICIPIO_RESIDENCIA','CO_UF_RESIDENCIA', 'SG_UF_RESIDENCIA',
                           'CO_MUNICIPIO_NASCIMENTO', 'SG_UF_NASCIMENTO', 'TP_ANO_CONCLUIU',
                           'TP_ENSINO', 'CO_UF_ESC', 'SG_UF_ESC', 'TP_DEPENDENCIA_ADM_ESC',
                                   'TP_LOCALIZACAO_ESC', 'TP_SIT_FUNC_ESC'])
```

4.corrigir o erro no número de casas nas notas (usar loc e dividir por 10)

```python
dados.loc[:,'NU_NOTA_CN'] /=10
dados.loc[:,'NU_NOTA_CH'] /=10
dados.loc[:,'NU_NOTA_LC'] /=10
dados.loc[:,'NU_NOTA_MT'] /=10
#os dois pontos indica todas as linhas da tabela
```

5.renomear as colunas para ficar mais fácil de chamá-las (.rename)

```python
dadosenem = dados.rename(columns = {'NU_NOTA_CN': 'NOTA_CN','NU_NOTA_CH': 'NOTA_CH',
                                    'NU_NOTA_LC': 'NOTA_LC','NU_NOTA_MT': 'NOTA_MT',
                                    'CO_ESCOLA': 'ESCOLA','NU_IDADE': 'IDADE',
                                    'TP_SEXO': 'SEXO', 'TP_COR_RACA': 'RACA',
                                    'Q025': 'INTERNET'})
```

```python
dadosenem.head()
```

```python
dadosenem['RACA'] = dadosenem['RACA'].replace({0:"não declarado", 1:"branco", 2: "preto", 3: "pardo", 4: "amarelo", 5: "indigena"})
```

```python
dadosenem['TP_LINGUA'] = dadosenem['TP_LINGUA'].replace([0,1], ["inglês", "espanhol"])
```

```python
dadosenem['ESCOLA'] = dadosenem['ESCOLA'].replace([1,2,3,4], ["não respondeu","publica","privada","exterior"])
```

```python
dadosenem['INTERNET'] = dadosenem['INTERNET'].replace(["A","B"], ["NÃO", "SIM"])
```

```python
dadosenem.head()
```

7.verificar as idades e excluir os menores de 11 anos (selecionando apenas os maiores de 11 anos)

```python
dadosenem['IDADE'].value_counts()
```

```python
dadosenem['IDADE'].value_counts().sort_index()
```

```python
dadosmaioresde11 = dadosenem.loc[dadosenem.IDADE >11]
```

```python
dadosmaioresde11.head()
```

8.excluir os treineiros (.loc)

```python
vestibulandos = dadosmaioresde11.loc[dadosmaioresde11.IN_TREINEIRO ==0]
```

9.selecionar apenas os que tiveram presença nos 2 dias

```python
vestibulandos['TP_PRESENCA_CN'].value_counts()
```

```python
vestibulandos['TP_PRESENCA_CH'].value_counts()
```

```python
vestibulandos['TP_PRESENCA_MT'].value_counts()
```

```python
vestibulandos['TP_PRESENCA_LC'].value_counts()
```

```python
vestibulandos_classificados = vestibulandos.loc[vestibulandos.TP_PRESENCA_LC ==1]
```

```python
vestibulandos_classificados = vestibulandos_classificados.loc[vestibulandos_classificados.TP_PRESENCA_MT ==1]
```

10.salvar a tabela tratada

```python
vestibulandos_classificados = vestibulandos_classificados.drop(columns=['IN_TREINEIRO'])
```

```python
#nome da tabela o formato (nome do arquivo, codifica, index-False para não salvar o índice junto '0,1,2,3,4...')
vestibulandos_classificados.to_csv('enem_2019_tratado.csv', encoding = 'iso-8859-1', index=False)
```
